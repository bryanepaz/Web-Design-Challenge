## Lat vs Temp

```python
cities['Lat'].max()
```

```python
#lat_temp = cities.drop(columns = ['City_ID', 'City', 'Cloudiness', 'Country', 'Date', 'Humidity', 'Lng', 'Wind Speed'])
lat_sort = cities.sort_values(['Lat'], ascending = True).reset_index(drop = True)
lat_sort.head()
```

```python
latitude = lat_sort['Lat'].values.tolist()
max_temp = lat_sort['Max Temp'].values.tolist()
```

```python
plt.scatter(latitude, max_temp, marker="o", facecolors="blue", alpha=0.50)

plt.title("Max Temp vs Latitude")
plt.xlabel("Latitude")
plt.ylabel("Max Temp (F)")
plt.xlim(-80, 100)
plt.ylim(-50, 150)

plt.figure(figsize=(400,278))
```
